```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101) 
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4) # we can change the number of centers 
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
data[0]
```

```python
data[1] # Original cluster
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6)) # 1 row 2 column

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
kmeans_model = KMeans(n_clusters=2)
```

```python
kmeans_model.fit(df.drop('Private',axis=1))
```

```python
df['Private']
```

```python
kmeans_model.labels_
```

```python
kmeans_model.cluster_centers_
```

```python
def converter(private):
  if private == 'Yes':
    return 1
  else:
    return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head(3)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(df['Cluster'],kmeans_model.labels_))
print('\n')
print(classification_report(df['Cluster'],kmeans_model.labels_))
```
